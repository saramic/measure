```python
import pandas as pd

df = pd.read_csv("chicks.csv")
df = df.rename(columns=lambda x: x.strip())
df = df.set_index("name")

df
```

```python
df_dates = df.drop(columns=['tag'])
df_dates = df_dates.rename(columns=lambda x: pd.to_datetime(x))
df_dates = df_dates.transpose()
df_dates
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")

plt.figure(figsize=(20,14))
plt.plot(df_dates)
plt.legend(df_dates.columns, loc='upper left')
plt.show()
```
