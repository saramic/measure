# Grapefruit Shrinkage

The analysis of shrinkage of stock by storage method, on home shelf vs home fridge.

```python
import pandas as pd

df = pd.read_csv("grapefruit.csv")
df = df.rename(columns=lambda x: x.strip())
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M')
df = df.set_index("date")

df
```

```python
df_dates = df
df_dates['shelf loss'] = (338.87 - df_dates["shelf"]) / 338.87 * 100
df_dates['fridge loss'] = (380.87 - df_dates["fridge"]) / 380.87 * 100
# better to NOT delete and just NOT plot them below
df_dates = df_dates.drop(columns=['shelf','fridge'])
df_dates
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")

plt.figure(figsize=(20,14))
plt.plot(df_dates)
plt.legend(df_dates, loc='upper left')
plt.show()
```
